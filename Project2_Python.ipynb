# 6. Check if there are missing values present in the column Rating, If any? drop them and and create a new column as 'Rating_category' by converting ratings to high and low categories(>3.5 is high rest low)

```python


# Check for missing values in a particular column
column_name = 'Rating'  # Replace 'column_name' with the actual column name you want to check

missing_values = dataframe[column_name].isnull().sum()

print(f"The column '{column_name}' has {missing_values} missing values.")
```

```python


# Drop rows with missing values in a Rating column
column_name = 'Rating'  # Replace 'column_name' with the actual column name
dataframe = dataframe.dropna(subset=[column_name])

# After dropping the rows, we need to reset the index
dataframe = dataframe.reset_index(drop=True)

# Print the updated DataFrame
dataframe
```

```python
dataframe["Rating_Category"] = dataframe["Rating"].apply(lambda x: "High" if x>3.5 else "Low")
dataframe
```
